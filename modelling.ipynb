```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv("./outputs/energy_data.csv")
df.drop("Unnamed: 0", axis=1, inplace=True)
df.head()
```

```python
hour_df = df.groupby(["start_date", "start_hour"]).sum().reset_index().drop(["start_minute", "end_hour", "end_minute"], axis=1)
```

```python
hour_df["consumption"].diff().plot()
```
